### Korrelation

Der Korrelationskoeffizient **r** gibt die Strke des **linearen** Zusammenhangs zwischen 2 features an

### Training

1. model erzeugen
2. model an Trainingsdaten fitten
3. model evaluieren

```python
X = housing[['median_income']]
y = housing.median_house_value
```

```python
sns.regplot(x=X, y=y, line_kws = {'color': 'red'})
```

### Linear Regression

Finden einer Gerade, welche den Fehler minimiert
